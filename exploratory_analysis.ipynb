```python
!pip install lifelines
import pandas as pd
import matplotlib.pyplot as plt
from lifelines import NelsonAalenFitter

# Load your dataset
df = pd.read_csv("/content/FINAL FILE.csv")  # Replace with your actual file path

# Define the time-to-event and event indicator columns
time_col = "duration"  # Column representing time until recidivism or censoring
event_col = "is_recid"  # Column indicating whether recidivism occurred (1 = Yes, 0 = No)

# Fit the Nelson-Aalen estimator
naf = NelsonAalenFitter()
naf.fit(df[time_col], event_observed=df[event_col])

# Plot the CIF (Cumulative Hazard Function)
plt.figure(figsize=(8, 5))
naf.plot(ci_show=True)
plt.title("Cumulative Incidence Function (Recidivism Risk Over Time)")
plt.xlabel("Time (Days)")
plt.ylabel("Cumulative Hazard (Risk of Recidivism)")
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset (replace 'YOUR_FILE.csv' with your actual file path)
df = pd.read_csv("/content/FINAL FILE.csv")

# Define survival time column (replace with actual column name)
time_col = "duration"  # Change this if your dataset has a different column name for survival time

# Set the figure size
plt.figure(figsize=(12, 6))

# Create a subplot with 1 row and 2 columns
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# Boxplot of survival time
sns.boxplot(y=df[time_col], ax=axes[0], color="skyblue")
axes[0].set_title("Boxplot of Survival Time")
axes[0].set_ylabel("Time to Recidivism (Days)")

# Violin plot of survival time
sns.violinplot(y=df[time_col], ax=axes[1], color="lightgreen")
axes[1].set_title("Violin Plot of Survival Time")
axes[1].set_ylabel("Time to Recidivism (Days)")

# Show the plots
plt.tight_layout()
plt.show()
```
